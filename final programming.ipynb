```python
x = df[['Age','Sex','Cp','Trtbps','Chol','Fbs','Restecg','Thalachh','Exng','Oldpeak','Slp','Caa','Thall']]
```

```python
y = df[['Output']]
```

```python
y=y.reshape(-1,1)
y
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#%% split training set to validation set
Xtrain, Xval, Ztrain, Zval = train_test_split(xtrain, ytrain, test_size=0.2, random_state=43)
```

```python
logreg = LogisticRegression()
logreg.fit(xtrain, ytrain)
acc_log = round(logreg.score(xtrain, ytrain) * 100, 2)
acc_log
ypred=logreg.predict(xtest)
```

acc_test_log = round(logreg.score(xtest, ytest) * 100, 2)

```python
ypred=logreg.predict(xtest)
dff=pd.DataFrame(data={'P V':ypred.flatten(), 'A V':ytest.flatten()})
dff
```

```python
decision_tree = DecisionTreeClassifier()
decision_tree.fit(xtrain, ytrain)
acc_decision_tree = round(decision_tree.score(xtrain, ytrain) * 100, 2)
acc_decision_tree
```

```python
acc_test_decision_tree = round(decision_tree.score(xtest, ytest) * 100, 2)
acc_test_decision_tree
```

```python
random_forest = GridSearchCV(estimator=RandomForestClassifier(), param_grid={'n_estimators': [100, 300]}, cv=5).fit(xtrain, ytrain)
random_forest.fit(xtrain, ytrain)
acc_random_forest = round(random_forest.score(xtrain, ytrain) * 100, 2)
print(acc_random_forest,random_forest.best_params_)
```

```python
acc_test_random_forest = round(random_forest.score(xtest, ytest) * 100, 2)
acc_test_random_forest
```

```python
models = pd.DataFrame({
    'Model': ['Logistic Regression',  
              'Decision Tree Classifier', 'Random Forest',  ],
    
    'Score_train': [acc_log, 
              acc_decision_tree, acc_random_forest ],
     'Score_test': [acc_test_log, 
              acc_test_decision_tree, acc_test_random_forest]
                    })
```

```python
models.sort_values(by=['Score_train', 'Score_test'], ascending=False)
```

```python
models.sort_values(by=['Score_test', 'Score_train'], ascending=False)
```

```python
models['Score_diff'] = abs(models['Score_train'] - models['Score_test'])
models.sort_values(by=['Score_diff'], ascending=True)
```

```python
plt.figure(figsize=[5,5])
xx = models['Model']
plt.tick_params(labelsize=10)
plt.plot(xx, models['Score_train'], label = 'Score_train')
plt.plot(xx, models['Score_test'], label = 'Score_test')
plt.legend()
plt.title('Score of 3 popular models for train and test datasets')
plt.xlabel('Models')
plt.ylabel('Score, %')
plt.xticks(xx, rotation='vertical')

plt.show()
```
